A quick Comparision of LASSO,Ridge and Linear Regression

```python
#Load Dataset
from sklearn.datasets import load_boston
```

```python
#import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=load_boston()
df
```

```python
# coverting the dataset into dataframe
df1=pd.DataFrame(df.data)
df1
```

```python
#changing the column names
df1.columns=df.feature_names
df1.head()
```

```python
#adding the target column
df1['price']=df.target
df1.head()
```

```python
# spling the data for dependent and independent data
X=df1.iloc[:,:-1] 
y=df1.iloc[:,-1] 
```

# Linear Regression

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
mse=cross_val_score(lr,X,y,scoring='neg_mean_squared_error',cv=5)
mse
```

```python
mean_mse=np.mean(mse)
mean_mse
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge=Ridge()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
ridge_regressor=GridSearchCV(ridge,parameters,scoring='neg_mean_squared_error',cv=5)
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
```

```python
lasso=Lasso()
parameters={'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20,30,35,40,45,50,55,100]}
lasso_regressor=GridSearchCV(lasso,parameters,scoring='neg_mean_squared_error',cv=5)

lasso_regressor.fit(X,y)
```

```python
print(lasso_regressor.best_params_)
print(lasso_regressor.best_score_)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
# testing our data
predicted_lasso=lasso_regressor.predict(x_test)
predicted_ridge=ridge_regressor.predict(x_test)
```

```python
sns.distplot(y_test-predicted_lasso)
```

```python
sns.distplot(y_test-predicted_ridge)
```
